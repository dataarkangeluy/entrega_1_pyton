```python
#################  Recursos Necesarios ##############
# !pip install unidecode  # Instala sistema de codificación de caracteres para levantar archivo



################# Bienvenida #################

print("""Bienvenido al juego del ahorcado..!

                               \|||/
                               (o o)
                      ------ooO-(_)-Ooo------
                       EL JUEGO DEL AHORCADO 

El Objetivo del juego es descubrir el nombre de la pelicula secreta.
Al iniciar el juego podras pedir una letra o adivinar la pelicula.
Si la letra no está en el nombre de la pelicula, entonces perderas una vida....

Logras la victoria si logras adivinar la pelicula secreata, antes de gastar tus 5 vidas.

MUCHA SUETE !!!  

""")


#################  LEO ARCHIVO - Cargo lista de peliculas desde archivo y limpio nombres de posible basura con encoding ##############
# abro el archivo 
with open('mis_pelis.txt', encoding="utf-8") as f:
    mis_peliculas = f.readlines()
f.close() 


################# LIMPIO LISTA - Cargo la lista con linenas de archivo y eliminar todos los espacios en blanco iniciales y finales de la cadena de txto que llega del archivo.
mis_peliculas_limpias = []  
for pelicula in mis_peliculas:
    mis_peliculas_limpias.append(pelicula.strip())  # eliminar espacios en blanco iniciales y finales de la cadena
# print(mis_peliculas_limpias)




################# Conecto e identifico el jugador 
nombre=input("Como te llamas? ")
print(f"hola, {nombre}, es hora de jugar !!")



#################  inicializacion de variables ##############
tus_letras=" "
vidas=5



################# PELICULA RANDOM - eligo la pelicula al azar y la muestro en pantalla al participante ##############
import random # importa el modulo random de python
def selectRandom(mis_peliculas_limpias):
    return random.choice(mis_peliculas_limpias)
pelicula_random = selectRandom(mis_peliculas_limpias)
palabra = [x.lower() for x in pelicula_random]  #convierto a minuculas la pelicula secreta

print(" ")
print(f"La pelicula secreta es: {pelicula_random}")





################# PRESENTO PISTAS - presento en pantalla la cantidad de letas y espacios que tiene la pelicula.

while vidas > 0: # Controla el numero de vidas
    fallas=0  
        

    ##################################################################################
    ### con la instruccion FOR reviso letra a letra y resto vidas si hay error  ######    
    for letra in palabra:  
        if letra in tus_letras:     # verifico si la letra esta en la letra de la pelicula
            print(letra,end="")     # si esta la letra la imprimo 
        else:            
            print(fallas)
            fallas+=1   
           # print("*",end=" ")  # si no esta, oculo letras y sumo una vida al contador       
        
    
    
    ##################################################################################
    # verifico si al salir del FOR, las fallas == 0 a cero el participante gano y corto el ciclo while
    if fallas==0:
        print("""
            felicidades, ganaste""")     
        break  # corto con el ciclo while
        
        
        
    ##################################################################################  
    print(" ")  
    tuletra=input("Introduce una letra: ")
    tus_letras+=tuletra    

    if tuletra not in palabra:
        vidas-=1
        print(f"""
                Letra no encontrada, ahora te quedan {vidas} vidas.
        """)
        print(f"tu lista de letras: {tus_letras} ") # muestro la lista de letras ingresadas
        
    if vidas== 0:
        print("Game Over!")

    
    #############################################################################
    
else:  # el else de while que se ejecuta una vez que la condición sea falsa
    print("gracias por participar")

            
```
